```python
from IPython.display import clear_output

def print_board(board,position,char_player):
    clear_output()
    if position-1 <= 2:
        list_board=list(board[1])
        list_board[(position-1)*4 + 1]=char_player
        board[1]=''.join(list_board)
    elif position-4 <= 2:
        list_board=list(board[4])
        list_board[(position-4)*4 + 1]=char_player
        board[4]=''.join(list_board)
    elif position-7 <= 2:
        list_board=list(board[7])
        list_board[(position-7)*4 + 1]=char_player
        board[7]=''.join(list_board)
        
    for item in board:
        print(item)
    return board



def converter(position):
    row=int(position/3)+1
    column=position%3
    if position%3==0:
        row=int(position/3)
        column=3
    return (row , column)

            
    
def check(positions):
    row=[]
    column=[]
    i=1
    
    if len(positions)>=3:
        
        old_positions=positions
        
        positions=list(map(converter,positions))
        
        # For columns and rows
        for item in positions:
            row.append(item[0])
            column.append(item[1])
        while i<=3:
            if row.count(i)==3:
                return True
            if column.count(i)==3:
                return True
            i+=1
        #For Diagonal
        counter=0
        possibilities=[(1,5,9),(3,5,7)]
        for item in possibilities:
            for num in item:
                if num in old_positions:
                    counter+=1
            if counter==3:
                return True
            else:
                counter=0
        return False          

    
    
play=True
game_chars=['X','O']
player1_positions=[]
player2_positions=[]
player1_char=''

while True:
    player1_char=(input('What are you going to play as X or O : ')).upper()
    if player1_char == 'X' or player1_char == 'O':
        break
    print('Your input is invalid please enter another one! ')     

game_chars.remove(player1_char)
player2_char=game_chars[0]

board=['   |   |   ',
       '   |   |   ',
       '___|___|___',
       '   |   |   ',
       '   |   |   ',
       '___|___|___',
       '   |   |   ',
       '   |   |   ',
       '   |   |   ']

empty_board=print_board(board,1,' ')

while play:
    while True:
        player1_pos=int(input('Player 1 what is your next location: '))
        if not(player1_pos in player2_positions or player1_pos in player1_positions):
            player1_positions.append(player1_pos)
            break
        print('This place is already taken!')
             
    new_board=print_board(board,player1_pos,player1_char)
    board=new_board
    
    if check(player1_positions):
        print('Player1 WINS Congrats!!!')
        break
        
    if len(player2_positions)+len(player1_positions)==9:
        print('You TIED!')
        break
        
    while True:
        player2_pos=int(input('Player 2 What is your next location: '))
        if not(player2_pos in player2_positions or player2_pos in player1_positions):
            player2_positions.append(player2_pos)
            break
        print('This place is already taken!')
    
    new_board=print_board(board,player2_pos,player2_char)
    board=new_board
    
    if check(player2_positions):
        print('Player2 WINS Congrats!!!')
        break
    
  
    
```
